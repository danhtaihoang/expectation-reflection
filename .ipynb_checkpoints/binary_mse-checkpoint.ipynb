## Synthetic data: Binary variables

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.metrics import accuracy_score

from synthesize_data import synthesize_data
import expectation_reflection as ER

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(1)
```

```python
l = 5000 ; n = 100 ; g = 0.5
```

```python
X,y,w0 = synthesize_data(l,n,g,data_type='binary')
```

```python
np.unique(y,return_counts=True)
```

```python
list_train_size = [0.8,0.6,0.4,0.2,0.1]
m = len(list_train_size)

w_pred = np.zeros((m,n))
mse = np.zeros(m)
accuracy_train = np.zeros(m)
accuracy_test = np.zeros(m)

for i,train_size in enumerate(list_train_size):
    X_train0,X_test,y_train0,y_test = train_test_split(X,y,test_size=0.2,random_state = 1)
    idx_train = np.random.choice(len(y_train0),size=int(train_size*len(y)),replace=False)
    X_train,y_train = X_train0[idx_train],y_train0[idx_train]
    h0,w = ER.fit(X_train,y_train,niter_max=300,regu=0.)

    mse[i] = ((w-w0)**2).mean()

    # w_pred
    w_pred[i,:] = w
    
    #train:
    y_pred_train = ER.predict(X_train,h0,w)
    accuracy_train[i] = accuracy_score(y_train,y_pred_train)

    # test
    y_pred_test = ER.predict(X_test,h0,w)
    accuracy_test[i] = accuracy_score(y_test,y_pred_test)
```

```python
plt.figure(figsize=(11,3.2))

plt.subplot2grid((1,3),(0,0))
plt.plot([-2,2],[-2,2],'r--')
plt.scatter(w0,w_pred[3,:],marker='^',color='r',label='train size = 0.1')
plt.scatter(w0,w_pred[0,:],marker='o',color='k',label='train size = 0.8')
plt.xlabel('actual interaction')
plt.ylabel('inferred interaction')
plt.legend()

plt.subplot2grid((1,3),(0,1))
plt.plot(list_train_size,mse,'ko-')
plt.xlabel('train size')
plt.ylabel('Mean Square Error')

plt.subplot2grid((1,3),(0,2))
plt.plot(list_train_size,accuracy_test,'ko-')
plt.xlabel('train size')
plt.ylabel('Predicted Accuracy')

plt.tight_layout(h_pad=1, w_pad=1.5)
plt.show()
```
