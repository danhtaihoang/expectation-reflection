```python
import sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pos_neg = np.loadtxt('pos_neg_equal_delta002.dat')
```

```python
nx,ny = 5,1
nfig = nx*ny
fig, ax = plt.subplots(ny,nx,figsize=(nx*3.5,ny*2.8))
   
xvalue = [0,1,2]
labels = ['DCA > ER','DCA ~ ER','DCA < ER']

for i,cutoff in enumerate([2, 3, 4, 5, 6]):
    ax[i].set_title('cutoff = %i $A^o$'%cutoff)
    ax[i].bar(0,pos_neg[2,i],color='k',width=0.8,label='DCA > ER')
    ax[i].bar(1,pos_neg[3,i],color='b',width=0.8,label='DCA ~ ER')
    ax[i].bar(2,pos_neg[1,i],color='r',width=0.8,label='DCA < ER')
    
    ax[i].set_xticks(xvalue)
    ax[i].set_xticklabels(labels)
       
    ax[i].set_ylabel('number of protein domains')

#ax[0].legend()

plt.tight_layout(h_pad=1, w_pad=1.5)
plt.savefig('fig5_pos_neg_equal.pdf', format='pdf', dpi=100)
```
